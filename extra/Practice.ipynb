```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
print(title)
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')

# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
for x in range(1, 8):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('span', class_='text')
    for quote in quotes:
        print('page:', x, '----------')
        print(quote.text)
    browser.links.find_by_partial_text('Next').click()
```

```python
url = 'http://books.toscrape.com/'
browser.visit(url)
```

```python
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
books = html_soup.find_all('article', class_='product_pod')
len(books)
```

```python
first_book = books[0]
first_book
```

```python
books_url = []
for book in books:
    link = book.h3.a.get('href')
    if link not in books_url:
        books_url.append(link)
```

```python
len(books_url)
```

```python
for book in books:
    book_link = book.h3.a.get('href')
    print(book_link)
```

```python
for x in range(1, 5):
    html = browser.html
    url_soup = soup(html, 'html.parser')
    books = url_soup.find_all('article', class_='product_pod')
    for book in books:
        book_title = book.h3.a.get('title')
        book_url = book.h3.a.get('href')
        print('page:', x, '----------')
        print(f'Book Title: {book_title}')
        print(f'Book URL: {book_url}')
    browser.links.find_by_partial_text('next').click()
```
